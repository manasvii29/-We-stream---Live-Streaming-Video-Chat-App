```python
import socket 
import time
import threading


a_recv_s=socket.socket()  # Creating a socket
a_recv_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1) # So that we can reuse the same port again
am_send_s=socket.socket()
am_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)
ad_send_s=socket.socket()
ad_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)

m_recv_s=socket.socket()
m_recv_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)
ma_send_s=socket.socket()
ma_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)
md_send_s=socket.socket()
md_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)

d_recv_s=socket.socket()
d_recv_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)
dm_send_s=socket.socket()
dm_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)
da_send_s=socket.socket()
da_send_s.setsockopt(socket.SOL_SOCKET , socket.SO_REUSEADDR, 1)

a_recv_port=2024   # Port for server to recieve packets
am_send_port=2025  # Port for server to send packets
ad_send_port=2026


m_recv_port=2027
ma_send_port=2028
md_send_port=2029

d_recv_port=2030
dm_send_port=2031
da_send_port=2032

ip=""  # Here IP will be taken dynamically

a_recv_s.bind((ip, a_recv_port))     # Binds the address with the port
am_send_s.bind((ip, am_send_port))
ad_send_s.bind((ip, ad_send_port))

m_recv_s.bind((ip, m_recv_port))
ma_send_s.bind((ip, ma_send_port))
md_send_s.bind((ip, md_send_port))

d_recv_s.bind((ip, d_recv_port))
dm_send_s.bind((ip, dm_send_port))
da_send_s.bind((ip, da_send_port))

a_recv_s.listen()  # Enables the server to accept connections
am_send_s.listen()
ad_send_s.listen()

m_recv_s.listen()
ma_send_s.listen()
md_send_s.listen()

d_recv_s.listen()
dm_send_s.listen()
da_send_s.listen()

a_recv_session, a_recv_addr = a_recv_s.accept() # Accepts the connection, returns the new connection and address of the other end
am_send_session, am_send_addr = am_send_s.accept()
ad_send_session, ad_send_addr = ad_send_s.accept()

m_recv_session, m_recv_addr = m_recv_s.accept()
ma_send_session, ma_send_addr = ma_send_s.accept()
md_send_session, md_send_addr = md_send_s.accept()

d_recv_session, d_recv_addr = d_recv_s.accept()
dm_send_session, dm_send_addr = dm_send_s.accept()
da_send_session, da_send_addr = da_send_s.accept()

def a_recv1():
    while True:
        data=a_recv_session.recv(10000000) # Recieve data from a client
        data1=data
        time.sleep(2)
        am_send_session.send(data) # Sends data to client
        ad_send_session.send(data1)

def m_recv1():
    while True:
        data=m_recv_session.recv(10000000)
        data1=data
        time.sleep(2)
        ma_send_session.send(data)
        md_send_session.send(data1)

def d_recv1(): 
    while True:
        data=d_recv_session.recv(10000000)
        data1=data
        time.sleep(2)
        dm_send_session.send(data)
        da_send_session.send(data1)

# Applying threading so every client gets the data(images) simultaneously
a1_recv=threading.Thread(target=a_recv1)
m1_recv=threading.Thread(target=m_recv1)
d1_recv=threading.Thread(target=d_recv1)

a1_recv.start()
m1_recv.start()
d1_recv.start()
```
